```python
import numpy as np
from numpy.linalg import norm
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def odefn(t, y, a, b):
    # problem: y' = [y[1], -y[0]]
    # solution: y = [sin(t), cos(t)]
    return np.array([y[1], -y[0]])

t_span = (0, 10)
args = (2, 0.5)

y0 = np.array([1, 0])
sol = solve_ivp(odefn, t_span, y0, t_eval=np.linspace(*t_span, 1000)[1:-1], args=args)
asol = np.array([args[0]*np.cos(sol.t), args[1]*np.sin(sol.t)]).T
```

```python
print('x RMS err:', np.sqrt(np.mean((sol.y[0] - asol[:, 0])**2)))
print('y RMS err:', np.sqrt(np.mean((sol.y[1] - asol[:, 1])**2)))
plt.scatter(sol.t, abs(sol.y[0] - asol[:, 0]), label='x')
plt.scatter(sol.t, abs(sol.y[1] - asol[:, 1]), label='y')
plt.title('Absolute Error')
plt.legend()
plt.show()
```

```python

plt.plot(sol.t, np.cos(sol.t), label='Ana-x')
plt.plot(sol.t, sol.y[0], label='Num-x')
plt.plot(sol.t, -np.sin(sol.t), '.', label='Ana-y')
plt.plot(sol.t, sol.y[1], label='Num-y')
plt.scatter(sol.t, abs(sol.y[0] - asol[:, 0]), label='Err x')
plt.scatter(sol.t, abs(sol.y[1] - asol[:, 1]), label='Err y')
# plt.xlim(0, 3)
# plt.ylim(-1.02, -0.9)
plt.legend()
plt.show()
```
